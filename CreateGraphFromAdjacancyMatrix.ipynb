Import Libaries

```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
print("Initial setup completed")
```

Create graph from Adjacancy Matrix

```python
# Load adjacancy matrix from csv
df = pd.read_csv("adjacancy_matrix.csv")

#Drop Column one as it contains headers
df = df.drop(df.columns[[0]], axis=1)

# Convert Dataframe to matrix
a = df.to_numpy()
print(a)
```

```python
# Create Graph
G = nx.Graph(a)
pos = nx.spring_layout(G)
plt.figure() 

# Here 0,1,2,3 labels would be replaced by shorts sybmols defined in variable stock_short_labels
nx.draw(G, pos, with_labels = True)
plt.draw()

print(nx.info(G))
```
