Import Libaries

```python
import networkx as nx
print("Initial setup completed")
```

Add nodes and edges to a graph

```python
#Define new graph
G1 = nx.Graph()

# Add one node at a time
G1.add_node("a")

# Add multiple nodes at a time
G1.add_nodes_from(["b","c"])

# Add an edge
G1.add_edge("a", "b")

print("Nodes of graph: ", G1.nodes())
print("Edges of graph: ", G1.edges())
```

```python
nx.draw_networkx(G1, with_labels = True)
```

Add edge and nodes are added automatically

```python
# Define new graph
G2 = nx.Graph()

# Add one edge at a time
G2.add_edge("x", "y")

# Add multiple edges at a time
G2.add_edges_from([("x","z"),("y","z")])

print("Nodes of graph: ", G2.nodes())
print("Edges of graph: ", G2.edges())
```

```python
nx.draw_networkx(G2, with_labels = True)
```

Graph with labelled edges

```python
cities_new = {"x":"Delhi", "y":"Mumbai", "z":"Kolkata", "a":"Chennai"}
H1 = nx.relabel_nodes(G2,cities_new)
nx.draw_networkx(H1)
```
