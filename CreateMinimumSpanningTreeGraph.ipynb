Import Libaries

```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
print("Initial setup completed")
```

Create Minimum Spanning Tree Graph

```python
# Load adjacancy matrix from csv
df = pd.read_csv("adjacancy_matrix.csv")
df = df.drop(df.columns[[0]], axis=1)
a = df.to_numpy()
#a = create_matrix(num_of_rows, num_of_cols)
print(a)
```

```python
# Create Graph
G = nx.Graph(a)
pos = nx.spring_layout(G)
plt.figure() 

# Here 0,1,2,3 labels would be replaced by shorts sybmols defined in variable stock_short_labels
nx.draw(G, pos, with_labels = True)
plt.draw()

print(nx.info(G))
```

```python
# Create Minimum Spanning Tree graph
labels = {0:'P1',1:'P2',2:'P3',3:'P4',4:'P5'}
T=nx.minimum_spanning_tree(G)
nx.draw_networkx(T, with_labels = True, labels=labels, node_size=50, width=0.3, font_size = 12)
```
